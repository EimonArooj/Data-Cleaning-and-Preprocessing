```python
import pandas as pd
Data=pd.read_csv('Applicants.csv')
Data
```

```python
#First I'll explore my dataset
Data.shape # checking shape
print('Missing Data\n:',Data.isnull().sum())# Data do contain missing values
#Also check duplicate values
print('Duplicate Data:\n',Data.duplicated().sum())
```

```python
#First we'll handle missing values
#First check dtypes of each col for filling
Data.dtypes
'''for categorical columns like unaiversity and status,,fill them with mode,,whil drop those rows with missing name/email as
its important data and without it,we cannot identify applicant'''
Data['University']=Data['University'].fillna('mode')
Data['Status']=Data['Status'].fillna('mode')
#Check whether these are filled are not
print('Missing Data\n:',Data.isnull().sum()) #Done
```

```python
#Now deal with name and email
#Drop that rows
Data = Data.dropna(subset=["Applicant Name"]) #where name is empty
Data = Data.dropna(subset=["Email"]) #where email is empty
print('Missing Data\n:',Data.isnull().sum())  #Done..no null data
```

```python
#now we drop dupicates first check no of dulicates adn dataset shape before duplication
print('Duplicate Data:\n',Data.duplicated().sum())
#now we drop duplicates first check no of dulicates adn dataset shape before duplicationheck shape before duplicantion
print('Dataset shape before duplication:',Data.shape)
df=Data.drop_duplicates()#cleaned dataset
#now check duplicates
print('Duplicate Data:\n',df.duplicated().sum())
#check shape now
print('Dataset shape after duplication:',df.shape) # 10 rows dropped
```

```python
#Standardize Text Fields
#=====================Applicant Name Analysis====================
## Count rows having extra spaces
print("\nRows with Leading/Trailing Spaces:")
print((df["Applicant Name"] != df["Applicant Name"].str.strip()).sum())
# Count different formats
print("\nDifferent Name Formats (Top 10):")
print(df["Applicant Name"].value_counts().head(10))
#removing rows
df["Applicant Name"] = df["Applicant Name"].str.strip()
df["Applicant Name"] = df["Applicant Name"].str.title()
#Now verify
print("\nRows with Leading/Trailing Spaces After Cleaning:")
print((df["Applicant Name"] != df["Applicant Name"].str.strip()).sum())
```

```python
#=====================Email column Analysis====================
#we haev to ensure email has no extra sapces and all emails shuld be in lowercase
#email containig upper case letters
print('Rows having upper case letter emails:')
print((df['Email']!=df['Email'].str.lower()).sum())
#email containig extra spaces
print('Rows having extra spaces:')
print((df['Email']!=df['Email'].str.strip()).sum())
#ALREADY CLEANED AND WELL FORMATTED
```

```python
#Phone numr standardization
print("\nPhone Numbers containing hyphens:")
print(df["Phone"].str.contains("-", na=False).sum())
print("\nPhone Numbers containing spaces:")
print(df["Phone"].str.contains(" ", na=False).sum())
#CLEANING
# Remove all spaces
df["Phone"] = df["Phone"].str.replace(" ", "", regex=False)
# Remove all hyphens
df["Phone"] = df["Phone"].str.replace("-", "", regex=False)
#VERIFY
print("\nPhone Numbers Containing Hyphens After Cleaning:")
print(df["Phone"].str.contains("-", na=False).sum())
print("\nPhone Numbers Containing Spaces After Cleaning:")
print(df["Phone"].str.contains(" ", na=False).sum())
```

```python
# ==========================================================
# DOMAIN APPLIED STANDARDIZATION
# ==========================================================
print("="*60)
print("DOMAIN APPLIED ANALYSIS")
print("="*60)

print("\nUnique Domain Values Before Cleaning:")
print(df["Domain Applied"].unique())

print("\nFrequency of Each Domain Before Cleaning:")
print(df["Domain Applied"].value_counts())

print("\nRows Containing Extra Spaces:")
print((df["Domain Applied"] != df["Domain Applied"].str.strip()).sum())

domain_mapping = {
    "data science": "Data Science",
    "datascience": "Data Science",
    "web development": "Web Development",
    "web dev": "Web Development",
    "mobile development": "Mobile Development",
    "mobile dev": "Mobile Development",
    "uiux design": "UI/UX Design",
    "ui ux design": "UI/UX Design",
    "ui/ux design": "UI/UX Design",
    "uiux": "UI/UX Design",
    "ui ux": "UI/UX Design",
    "ui/ux": "UI/UX Design",
    "cybersecurity": "Cyber Security",
    "cyber security": "Cyber Security"
}

print("\nFormats Present in Dataset and Their Standard Form:\n")
unique_domains = df["Domain Applied"].unique()
for value in unique_domains:
    cleaned_value = value.strip().lower()
    if cleaned_value in domain_mapping:
        print(f"{value:<25} ---> {domain_mapping[cleaned_value]}")
    else:
        print(f"{value:<25} ---> {value}")

df["Domain Applied"] = df["Domain Applied"].str.strip()
df["Domain Applied"] = df["Domain Applied"].str.lower()
df["Domain Applied"] = df["Domain Applied"].replace(domain_mapping)

print("\nUnique Domain Values After Cleaning:")
print(df["Domain Applied"].unique())

print("\nFrequency After Cleaning:")
print(df["Domain Applied"].value_counts())

print("\nRows Containing Extra Spaces After Cleaning:")
print((df["Domain Applied"] != df["Domain Applied"].str.strip()).sum())
```

```python
#University column
# ==========================================================
# University Standardization
# ==========================================================

print("="*60)
print("UNIVERSITY ANALYSIS")
print("="*60)

print("\nUnique Universities Before Cleaning:")
print(df["University"].unique())

print("\nFrequency Count:")
print(df["University"].value_counts())

print("\nRows containing extra spaces:")
print((df["University"] != df["University"].str.strip()).sum())

# ---------------- Cleaning ----------------

df["University"] = df["University"].str.strip()
df["University"] = df["University"].str.title()

# ---------------- Verification ----------------

print("\nUnique Universities After Cleaning:")
print(df["University"].unique())

print("\nFrequency After Cleaning:")
print(df["University"].value_counts())

print("\nRows containing extra spaces After Cleaning:")
print((df["University"] != df["University"].str.strip()).sum())
```

```python
# ==========================================================
# STATUS STANDARDIZATION
# ==========================================================
print("="*60)
print("STATUS ANALYSIS")
print("="*60)

print("\nUnique Status Values Before Cleaning:")
print(df["Status"].unique())

print("\nFrequency of Each Status Before Cleaning:")
print(df["Status"].value_counts(dropna=False))

print("\nRows Containing Extra Spaces:")
print((df["Status"] != df["Status"].str.strip()).sum())

# ---------------- FORMATS FOUND ----------------
print("\nFormats Present in Dataset and Their Standard Form:\n")
for value in df["Status"].dropna().unique():
    standard = value.strip().upper()
    print(f"{value:<20} ---> {standard}")

# ---------------- CLEANING ----------------
df["Status"] = df["Status"].str.strip()
df["Status"] = df["Status"].str.upper()

# ---------------- VERIFICATION ----------------
print("\nUnique Status Values After Cleaning:")
print(df["Status"].unique())

print("\nFrequency After Cleaning:")
print(df["Status"].value_counts(dropna=False))

print("\nRows Containing Extra Spaces After Cleaning:")
print((df["Status"] != df["Status"].str.strip()).sum())
```

```python
#convert and fix dtype of date column
# ==========================================================
# CONVERT AND FIX DATA TYPES (Application Date)
# ==========================================================
print("="*60)
print("APPLICATION DATE - TYPE CONVERSION")
print("="*60)

print("\nData type before conversion:")
print(df["Application Date"].dtype)

print("\nSample values before conversion:")
print(df["Application Date"].head(5))

# Convert to actual datetime
df["Application Date"] = pd.to_datetime(df["Application Date"], format="%m/%d/%Y", errors="coerce")

print("\nData type after conversion:")
print(df["Application Date"].dtype)

print("\nSample values after conversion:")
print(df["Application Date"].head(5))

print("\nRows where conversion failed (became NaT):")
print(df["Application Date"].isna().sum())
```

```python
#Verify the final cleaned dataset
# ==========================================================
# VERIFY THE CLEANED DATASET
# ==========================================================
print("="*60)
print("FINAL VERIFICATION")
print("="*60)

print("\nDataset Shape:", df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values per Column:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nFirst 5 Rows:")
print(df.head())
```

```python
#Export Cleaned Dataset
# ==========================================================
# EXPORT THE CLEANED DATASET
# ==========================================================
df.to_csv("Cleaned_Applicants.csv", index=False)
print("Cleaned dataset exported successfully as 'Cleaned_Applicants.csv'")
```
